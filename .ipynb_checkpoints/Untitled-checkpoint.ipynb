```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from trading_simulator import TradingSimulator
```

```python
prices = pd.read_csv('prices.csv', index_col='Date', parse_dates=['Date'])
ret = prices.diff()

equity_ccy = {
    'Stock_01': 'Crncy_03', 'Stock_02': 'Crncy_04', 'Stock_03': 'Crncy_04',
    'Stock_04': 'Crncy_02', 'Stock_05': 'Crncy_03', 'Stock_06': 'Crncy_02',
    'Stock_07': 'Crncy_03', 'Stock_08': 'Crncy_02', 'Stock_09': 'Crncy_04',
    'Stock_10': 'Crncy_03', 'Stock_11': 'Crncy_01', 'Stock_12': 'Crncy_04',
    'Stock_13': 'Crncy_01', 'Stock_14': 'Crncy_01', 'Stock_15': 'Crncy_01',
}

fx_pairs_map = {
    'FX_01': ('Crncy_02', 'Crncy_01'),
    'FX_02': ('Crncy_04', 'Crncy_02'),
    'FX_03': ('Crncy_04', 'Crncy_03'),
    'FX_04': ('Crncy_02', 'Crncy_03'),
    'FX_05': ('Crncy_01', 'Crncy_03'),
    'FX_06': ('Crncy_04', 'Crncy_01'),
}

log_ret = np.log(prices).diff()
#log_ret = np.log(prices.resample('W-FRI').last()).diff()
```

```python
plt.figure(figsize=(12, 10))
plt.imshow(log_ret.corr(), aspect='auto')
plt.colorbar()
plt.xticks(range(len(log_ret.columns)), log_ret.columns, rotation=90)
plt.yticks(range(len(log_ret.columns)), log_ret.columns)
plt.tight_layout()
plt.show()
```

```python
# plot returns
plt.figure(figsize=(10,5))
plt.plot(log_ret.index, log_ret['FX_01'], label='FX_01')
plt.plot(log_ret.index, log_ret['FX_04'], label='FX_04')
plt.legend()
plt.tight_layout()
plt.show()

# spread
spread = log_ret['FX_01'] - log_ret['FX_04']

# plot spread
plt.figure(figsize=(10,5))
plt.plot(spread.index, spread)
plt.tight_layout()
plt.show()
```

```python
x = log_ret['FX_04']
y = log_ret['FX_01']

beta = (y.rolling(30).cov(x) / x.rolling(30).var()).replace([np.inf, -np.inf], np.nan)

residual = (y - beta * x)
spread = residual.cumsum()

spread.plot()

z = ((spread - spread.rolling(30).mean()) / spread.rolling(30).std()).fillna(0)

target_shares_df = pd.DataFrame(0, index=x.index, columns=prices.columns)
target_shares_df['FX_01'] = np.where(z > 1.5, -5000, np.where(z < -1.5, 5000, 0))
target_shares_df['FX_04'] = -target_shares_df['FX_01'] * beta.fillna(1)

def strategy(row_pos, cash, portfolio, signal_prices, data):
    date = data.index[row_pos]
    if date not in target_shares_df.index:
        return []
    return [
        ('BUY' if (tgt := int(target_shares_df.loc[date, t])) - portfolio.get(t, 0) > 0 else 'SELL',
         t, abs(tgt - portfolio.get(t, 0)))
        for t in ['FX_01', 'FX_04']
        if (tgt := int(target_shares_df.loc[date, t])) != portfolio.get(t, 0)
    ]

simulator = TradingSimulator(
    assets=list(prices.columns),
    initial_cash=100_000,
    equity_currency_map=equity_ccy,
    fx_pairs_map=fx_pairs_map,
)
simulator.run(strategy, prices, prices)
simulator.save_results("orders.csv", "portfolio.csv")
simulator.plot_performance(prices, save_file="performance_plot.png")
```
