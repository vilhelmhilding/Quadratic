```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
prices = pd.read_csv('prices.csv', index_col='Date', parse_dates=['Date'])

# Select stocks and ETF
stocks = prices[[f"Stock_{i:02d}" for i in range(1, 16)]]
etf = prices["Idx_04"]

# Reconstruct ETF (equal-weight, no rebalancing)
weights = np.ones(stocks.shape[1]) / stocks.shape[1]

# Normalize to 1 at start (important)
norm_stocks = stocks / stocks.iloc[0]

# Buy-and-hold portfolio (no rebalancing)
synthetic_etf = (norm_stocks * weights).sum(axis=1)

# Scale to same base as Idx_04
synthetic_etf = synthetic_etf * etf.iloc[0]

# Discount / mispricing
discount = (etf - synthetic_etf) / synthetic_etf

# Plot ETF vs synthetic
plt.figure(figsize=(10,6))
plt.plot(etf, label="Idx_04 (ETF)")
plt.plot(synthetic_etf, label="Synthetic ETF", linestyle="--")
plt.legend()
plt.title("ETF vs Synthetic (No Rebalancing)")
plt.show()

# Plot discount
plt.figure(figsize=(10,4))
plt.plot(discount, label="Discount")
plt.axhline(0, linestyle="--")
plt.legend()
plt.title("ETF Mispricing (Discount)")
plt.show()
```

```python
# Currency denomination for each equity
equity_ccy = {
    'Stock_01': 'Crncy_03', 'Stock_02': 'Crncy_04', 'Stock_03': 'Crncy_04',
    'Stock_04': 'Crncy_02', 'Stock_05': 'Crncy_03', 'Stock_06': 'Crncy_02',
    'Stock_07': 'Crncy_03', 'Stock_08': 'Crncy_02', 'Stock_09': 'Crncy_04',
    'Stock_10': 'Crncy_03', 'Stock_11': 'Crncy_01', 'Stock_12': 'Crncy_04',
    'Stock_13': 'Crncy_01', 'Stock_14': 'Crncy_01', 'Stock_15': 'Crncy_01',
}

# FX pairs: ticker → (base_currency, quote_currency)
fx_pairs_map = {
    'FX_01': ('Crncy_02', 'Crncy_01'),
    'FX_02': ('Crncy_04', 'Crncy_02'),
    'FX_03': ('Crncy_04', 'Crncy_03'),
    'FX_04': ('Crncy_02', 'Crncy_03'),
    'FX_05': ('Crncy_01', 'Crncy_03'),
    'FX_06': ('Crncy_04', 'Crncy_01'),
}
```

```python
# --- bygg valutalog-returns från FX (log-return space) ---
ccy = sorted(set(sum(fx_pairs_map.values(), ())))
base = ccy[0]

ccy_ret = pd.DataFrame(index=log_ret.index)
ccy_ret[base] = 0.0

for _ in range(len(ccy)):
    for fx, (b, q) in fx_pairs_map.items():
        if b in ccy_ret and q not in ccy_ret:
            ccy_ret[q] = ccy_ret[b] - log_ret[fx]
        if q in ccy_ret and b not in ccy_ret:
            ccy_ret[b] = ccy_ret[q] + log_ret[fx]

# --- lägg till i original df ---
log_ret = pd.concat([log_ret, ccy_ret.add_prefix('CCY_')], axis=1)

print(log_ret)

(ccy_ret.cumsum()).plot(figsize=(10,6))
plt.show()
```

```python
stocks = list(equity_ccy.keys())

log_ret_base = log_ret[stocks].copy()

for s in stocks:
    ccy = equity_ccy[s]
    log_ret_base[s] = log_ret[s] + ccy_ret[ccy]

log_ret_base.cumsum().plot(figsize=(10,6))
plt.show()
```

```python
log_ret_base.cumsum()[["Stock_09", "Stock_11"]].plot(figsize=(10,6))
plt.show()
```

```python
import statsmodels.api as sm

stocks = list(equity_ccy.keys())
idx = log_ret["Idx_04"]

# regress bort Idx_04
resid = pd.DataFrame(index=log_ret_base.index)

for s in stocks:
    df = pd.concat([log_ret_base[s], idx], axis=1).dropna()
    X = sm.add_constant(df["Idx_04"])
    y = df[s]
    resid.loc[df.index, s] = sm.OLS(y, X).fit().resid

# plot residualer (kumulativa)
resid.cumsum().plot(figsize=(10,6))
plt.show()

# korrelationsmatris
corr = resid.corr()
print(corr)

import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(resid.corr(), annot=False, cmap="coolwarm", center=0)
plt.show()
```
